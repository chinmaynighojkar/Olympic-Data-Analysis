```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load dataset
athletes = pd.read_csv('C:\Projects\Olympic Data Analysis')
region = pd.read_csv('C:\Projects\Olympic Data Analysis\noc_regions.csv')
```

```python
athletes.head()
```

```python
region.head()
```

```python
#joining the dataframes

athletes_df = athletes.merge(region, how='left', on='NOC')
athletes_df.head()
```

```python
athletes_df.shape
```

```python
#making data frame more aesthetic

athletes_df.rename(columns= {'region':'Region','notes':'Notes'}, inplace='True')
```

```python
athletes_df.head()
```

```python
athletes_df.info()
```

```python
athletes_df.describe()
```

```python
null_values = athletes_df.isnull()
null_columns = null_values.any()
null_columns
```

```python
#tolal no of null values
athletes_df.isnull().sum()
```

```python
#Details of India
athletes_df.query('Team == "India"').head()
```

```python
#Japan Details
athletes_df.query('Team=="Japan"').head()
```

```python
#top countries participating
top_10_countries = athletes_df.Team.value_counts().sort_values(ascending = False).head(10)
top_10_countries
```

```python
#bar graph for top 10 countries 
plt.figure(figsize = (10,5))
plt.title('Top countries participating')
sns.barplot(x=top_10_countries.index, y=top_10_countries);
```

```python
#Age distribution of the participants
plt.figure(figsize=(10,5))
plt.title("Age distribution of athletes")
plt.xlabel('Age')
plt.ylabel('Number of Participants')
plt.hist(athletes_df.Age, bins = np.arange(10,80,2),edgecolor='white', color='green');
```

```python
#male and female participants
gender_count= athletes_df.Sex.value_counts()
gender_count
```

```python
#pie plot for male and female participants
plt.figure(figsize=(10,5))
plt.title("Gender Distribution")
plt.pie(gender_count, labels=gender_count.index,startangle=180)
plt.show()
```

```python
#total medals
medal_count= athletes_df.Medal.value_counts()
medal_count
```

```python
#total number of female in olympics by year
womenolympics=athletes_df[(athletes_df.Sex=='F')& (athletes_df.Season=='Summer')][['Sex','Year']]
womenolympics=womenolympics.groupby('Year').count().reset_index()
womenolympics.tail() #for more recent data
```

```python
female_participation=athletes_df[(athletes_df.Sex=='F')& (athletes_df.Season=='Summer')]
female_participation
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x='Year', data=female_participation)
plt.title('Women Participation')
plt.show()
```

```python
#gold beyond 60 years of age
goldmedal= athletes_df[(athletes_df.Medal=='Gold')]
goldmedal.head()
```

```python
goldmedal['ID'][goldmedal['Age']>60].count()
```

```python
sporting_event = goldmedal['Sport'][goldmedal['Age']>60]
sporting_event
```

```python
#plotting for sporting_event
plt.figure(figsize=(10,5))
sns.countplot(x=sporting_event, data=athletes_df)
plt.title('Gold medals for athletes above 60 years of age')
plt.show()
```

```python
#gold medals from each country
goldmedal.Region.value_counts().reset_index(name='Medal').head()
```

```python
golden= goldmedal.Region.value_counts().reset_index(name='Medal').head(6)
z= sns.catplot(x="index", y="Medal", data=golden, kind='bar', palette='rocket')
z.set_xlabels('top 6 countries')
z.set_ylabels('number of medals')
plt.title('gold medals per country')
```

```python
#height vs weight chart of olympic medalists
#excluding null values in medal column
not_nullmedals= athletes_df[(athletes_df['Height'].notnull()) & (athletes_df['Weight'].notnull())]
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(x='Height', y='Weight', data=not_nullmedals, hue='Sex')
plt.title('Height vs Weight of olympic medalists')
plt.show()
```
